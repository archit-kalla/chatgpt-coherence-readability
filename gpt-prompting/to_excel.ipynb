```python
import pandas as pd
from prompt import OUTPUT_FILE

OUTPUT_WORKBOOK = "results.xlsx"
DOMAINS_FILE = "domains.md"
```

```python
contents = open(OUTPUT_FILE, 'r').read()  # read in file
contents = contents.strip().split("### ") # split by header
contents = contents[1:]  # remove blank first entry

# separate prompts from output
contents_dict = {
    'prompt': [],
    'output': []
}

for i in range(len(contents)):
    newline_loc = contents[i].find('\n')
    prompt = contents[i][:newline_loc].strip()
    output = contents[i][newline_loc:].strip()
    contents_dict['prompt'].append(prompt)
    contents_dict['output'].append(output)
    
content_df = pd.DataFrame.from_dict(contents_dict)

content_df
```

```python
# add content headers
domain = ['none'] * len(content_df)
content_df.insert(0, 'domain', domain)
```

```python
content_df
```

```python
# load in domain classifications for prompts
contents = open(DOMAINS_FILE, 'r').read()  # read in file
contents = contents.strip().split("### ") # split by header
contents = contents[1:]  # remove blank first entry

# separate prompts from output
domains = {}

for i in range(len(contents)):
    newline_loc = contents[i].find('\n')
    domain = contents[i][:newline_loc].strip()
    prompts = contents[i][newline_loc:].strip().split('\n')
    domains[domain] = prompts
    
domains
```

```python
# simple dict classifier
def find_domain(prompt: str):
    for domain in domains:
        for p in domains[domain]:
            if p == prompt:
                return domain
    return None
```

```python
# classify domains in data frame
content_df.domain = content_df.prompt.map(find_domain)
content_df
```

```python
# write out our dataframe
content_df.to_excel(OUTPUT_WORKBOOK, index=False)
```

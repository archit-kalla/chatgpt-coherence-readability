```python
from sgnlp.models.coherence_momentum import CoherenceMomentumModel, CoherenceMomentumConfig, \
    CoherenceMomentumPreprocessor

# Load Model
config = CoherenceMomentumConfig.from_pretrained(
    "https://storage.googleapis.com/sgnlp-models/models/coherence_momentum/config.json"
)
model = CoherenceMomentumModel.from_pretrained(
    "https://storage.googleapis.com/sgnlp-models/models/coherence_momentum/pytorch_model.bin",
    config=config
)

preprocessor = CoherenceMomentumPreprocessor(config.model_size, config.max_len)
```

```python
# Example text inputs
text1 = "Companies listed below reported quarterly profit substantially different from the average of analysts ' " \
        "estimates . The companies are followed by at least three analysts , and had a minimum five-cent change in " \
        "actual earnings per share . Estimated and actual results involving losses are omitted . The percent " \
        "difference compares actual profit with the 30-day estimate where at least three analysts have issues " \
        "forecasts in the past 30 days . Otherwise , actual profit is compared with the 300-day estimate . " \
        "Source : Zacks Investment Research"
text2 = "The companies are followed by at least three analysts , and had a minimum five-cent change in actual " \
        "earnings per share . The percent difference compares actual profit with the 30-day estimate where at least " \
        "three analysts have issues forecasts in the past 30 days . Otherwise , actual profit is compared with the " \
        "300-day estimate . Source : Zacks Investment Research. Companies listed below reported quarterly profit " \
        "substantially different from the average of analysts ' estimates . Estimated and actual results involving " \
        "losses are omitted ."

text1_tensor = preprocessor([text1])
text2_tensor = preprocessor([text2])

text1_score = model.get_main_score(text1_tensor["tokenized_texts"]).item()
text2_score = model.get_main_score(text2_tensor["tokenized_texts"]).item()

print(text1_score, text2_score)
```

```python
# read in our data from GPT prompts
import pandas as pd
RESULTS_FILE = "../gpt-prompting/results.xlsx"
```

```python
results_df = pd.read_excel(RESULTS_FILE)
results_df
```

```python
def get_coherence(output: str):
    curr_tensor = preprocessor([output.strip()])
    score = model.get_main_score(curr_tensor["tokenized_texts"]).item()
    return score

coherence_scores = results_df.output.map(get_coherence)
coherence_scores
```

```python
results_df.insert(3, 'hf_coherence', coherence_scores)
results_df
```

```python
# some analysis
prompt_coherence = {}
domain_coherence = {}
for item in results_df.itertuples():
    domain_prompt_encoded = item.domain + "|" + item.prompt
    if domain_prompt_encoded not in prompt_coherence:
        prompt_coherence[domain_prompt_encoded] = [item.hf_coherence]
    else:
        prompt_coherence[domain_prompt_encoded].append(item.hf_coherence)
        
    if item.domain not in domain_coherence:
        domain_coherence[item.domain] = [item.hf_coherence]
    else:
        domain_coherence[item.domain].append(item.hf_coherence)
        
prompt_mean_coherence = {prompt: (sum(vals) / len(vals)) for (prompt, vals) in prompt_coherence.items()}
prompt_mean_coherence = pd.DataFrame.from_dict({
    'domain': [i.split('|')[0] for i in prompt_mean_coherence.keys()], 
    'prompt': [i.split('|')[1] for i in prompt_mean_coherence.keys()], 
    'avg_hf_coherence': prompt_mean_coherence.values()
})
prompt_mean_coherence = prompt_mean_coherence.sort_values('avg_hf_coherence', ascending=False)
prompt_mean_coherence
```

```python
domain_mean_coherence = {domain: (sum(vals) / len(vals)) for (domain, vals) in domain_coherence.items()}
domain_mean_coherence = pd.DataFrame.from_dict({'domain': domain_mean_coherence.keys(), 'avg_hf_coherence': domain_mean_coherence.values()})
domain_mean_coherence = domain_mean_coherence.sort_values('avg_hf_coherence', ascending=False)
domain_mean_coherence
```

```python
# write out all data to excel sheet
with pd.ExcelWriter(RESULTS_FILE) as writer:
    results_df.to_excel(writer, sheet_name="Results", index=False)
    prompt_mean_coherence.to_excel(writer, sheet_name="Prompt Avg. Coherence", index=False)
    domain_mean_coherence.to_excel(writer, sheet_name="Domain Avg. Coherence", index=False)
```
